```python
import pandas as pd
import re
import string
import os

def clean_text(text):
    """Clean text for emotion classification"""
    
    if pd.isna(text):
        return ""
    
    text = str(text)
    
    # 1. Convert to lowercase
    text = text.lower()
    
    # 2. Remove [NAME] and [RELIGION] placeholders
    text = re.sub(r'\[name\]|\[religion\]', '', text)
    
    # 3. Remove URLs
    text = re.sub(r'http\S+|www\S+|https\S+', '', text, flags=re.MULTILINE)
    
    # 4. Remove user mentions (@username) - ADD THIS
    text = re.sub(r'@\w+', '', text)
    
    # 5. Remove subreddit references (r/subreddit) - ADD THIS
    text = re.sub(r'r/\w+', '', text)
    
    # 6. Expand common contractions - ADD THIS (helps with meaning)
    contractions = {
        "n't": " not",
        "'s": " is",
        "'re": " are",
        "'ve": " have",
        "'ll": " will",
        "'m": " am",
        "'d": " would",
        "won't": " will not",
        "can't": " cannot",
        "don't": " do not",
        "doesn't": " does not",
        "didn't": " did not",
        "wasn't": " was not",
        "weren't": " were not",
        "isn't": " is not",
        "aren't": " are not",
    }
    for contraction, expansion in contractions.items():
        text = text.replace(contraction, expansion)
    
    # 7. Remove punctuation
    text = text.translate(str.maketrans('', '', string.punctuation))
    
    # 8. Remove numbers
    text = re.sub(r'\d+', '', text)
    
    # 9. Remove extra whitespace
    text = " ".join(text.split())
    
    return text


# Load your datasets
new_directory = 'C:/Users/User/Downloads/project_1'
os.chdir(new_directory)

train_df = pd.read_csv('data/train.csv')
valid_df = pd.read_csv('data/valid.csv')
test_df = pd.read_csv('data/test_no_label.csv')

print("Original sample:")
print(train_df['text'].iloc[0])
print("\n" + "="*50)

# Apply cleaning
print("\nCleaning data...")
train_df['cleaned_text'] = train_df['text'].apply(clean_text)
valid_df['cleaned_text'] = valid_df['text'].apply(clean_text)
test_df['cleaned_text'] = test_df['text'].apply(clean_text)

print("\nCleaned sample:")
print(train_df['cleaned_text'].iloc[0])

# Save cleaned data
train_df[['id', 'cleaned_text', 'label']].to_csv('data/train_cleaned.csv', index=False)
valid_df[['id', 'cleaned_text', 'label']].to_csv('data/valid_cleaned.csv', index=False)
test_df[['id', 'cleaned_text']].to_csv('data/test_cleaned.csv', index=False)

print("\n✅ Saved cleaned files:")
print("   - data/train_cleaned.csv")
print("   - data/valid_cleaned.csv")
print("   - data/test_cleaned.csv")
```
