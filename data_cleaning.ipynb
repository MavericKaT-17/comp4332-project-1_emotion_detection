```python
import pandas as pd
import re
import string
import os
```

```python
def clean_text(text):
    # 1. Convert to lowercase
    text = text.lower()
    
    # 2. Remove [NAME] and [RELIGION] placeholders if necessary
    # These are common in your specific dataset (e.g., train.csv)
    text = re.sub(r'\[name\]|\[religion\]', '', text)
    
    # 3. Remove URLs
    text = re.sub(r'http\S+|www\S+|https\S+', '', text, flags=re.MULTILINE)
    
    # 4. Remove punctuation (keeping some if you want to capture emojis later)
    text = text.translate(str.maketrans('', '', string.punctuation))
    
    # 5. Remove numbers
    text = re.sub(r'\d+', '', text)
    
    # 6. Remove extra whitespace
    text = " ".join(text.split())
    
    return text
```

```python
# Load your datasets
train_df = pd.read_csv('data/train.csv')
valid_df = pd.read_csv('data/valid.csv')
test_df = pd.read_csv('data/test_no_label.csv')
```

```python
# Apply cleansing
train_df['cleaned_text'] = train_df['text'].apply(clean_text)
valid_df['cleaned_text'] = valid_df['text'].apply(clean_text)
test_df['cleaned_text'] = test_df['text'].apply(clean_text)

# Example output
print("Original:", train_df['text'].iloc[0])
print("Cleaned:", train_df['cleaned_text'].iloc[0])

# Save the cleaned data for your model
os.makedirs('cleaned_data', exist_ok=True)
train_df[['id', 'cleaned_text', 'label']].to_csv('cleaned_data/train_cleaned.csv', index=False)
```
